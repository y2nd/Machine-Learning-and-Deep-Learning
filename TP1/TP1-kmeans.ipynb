K-Means is an algorithm that takes aims at clustering a given dataset into $k$ groups (called clusters) of data similar to each other. Each cluster is characterized by its centroid. 
In this practice, you are going to implement the kmeans algorithm and apply it on different dataset.

```python
def kmeans(dataset, k, max_iterations=5):
    
    centroids = get_random_centroids(dataset, k)
    iterations = 0
    old_centroids = None
    
    while not should_stop(old_centroids, centroids, iterations, max_iterations):
        old_centroids = centroids
        iterations += 1
        labels = get_labels(dataset, centroids)
        centroids = get_centroids(dataset, labels, k)
    
    return centroids, labels
```

```python
# Initialize centroids by choosing randomly k points from the dataset
def get_random_centroids(dataset, num_clusters):
    num_points, _ = dataset.shape
    centroids = dataset[np.random.randint(num_points, size=num_clusters), :]
    return centroids
```

```python
# Returns True or False if k-means if the maximum number of iterations is reached 
# or if the centroids do not change anymore
def should_stop(old_centroids, centroids, iterations, max_iterations):
        return ((iterations == max_iterations) or np.array_equal(old_centroids, centroids))
```

```python
# Returns the distance between two points
def get_distance(point1,point2):
    return np.linalg.norm(point1-point2)
```

```python
# Returns the label for each point in the dataSet. The label is the one of the closest centroid
def get_labels(dataset, centroids):
    numPoints, _ = dataset.shape
    labels = np.zeros(numPoints)
    
    for i in range(numPoints) : 
        labels[i] = np.argmin([get_distance(dataset[i], c) for c in centroids])
        
    return labels.astype(np.int)
```

```python
# Returns the centroids of the clusters. Each centroid is the geometric mean of the points that
# have that centroid's label. Important: If a centroid is empty (no points have
# that centroid's label) you should randomly re-initialize it.
def get_centroids(dataset, labels, k):
    num_points, _ = np.shape(dataset)
    A = [[] for i in range(k)]
    for i in range(num_points):
        A[labels[i]].append(dataset[i])

    centroids = []
    for i in range(k):
        if len(A[i]) == 0:
            centroids.append(dataset[np.random.randint(num_points),:])
        else:
            centroids.append(np.mean(np.vstack(A[i]), axis=0))

    return np.array(centroids)
```

#### Scikit-learn library
The Scikit-learn library proposes many functions for machine learning. You are going to compare the results obtained by your implementation with the ones from this library. 

* Study the documentation of the KMeans function and apply it to the 2D dataset.
* Compare

```python
from sklearn.cluster import KMeans

kmean_instance = KMeans(n_clusters=4, init='random')
result = kmean_instance.fit(X)
centers = result.cluster_centers_
labels = result.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')

plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

Les deux implémentations de Kmeans ne donnent pas toujours le même résultat. En effet,  si on exécute plusieurs fois notre algorithme on peut obtenir des résultats différents, parce que le point de départ de l’algorithme est le tirage aléatoire des centres des groupes. Donc si on tire aléatoirement de nouvelles données, les résultats peuvent être différents. Par contre, l'algorithme implémenté par la bibliothèque *sickit-learn* s'execute sur plusieurs cas initiaux (un paramètre n_init). Le résultat final correspond à celui le plus fréquent.
